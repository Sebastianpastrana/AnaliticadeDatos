```python
#Eliminar primera fila o filas innecesariaas
data1=pd.read_csv("inmuebles2.csv",skiprows=1)
data1
```

```python
#Eliminar columnas innecesarias
data2= data1.drop(['0','Referencia'],axis=1)
data2
```

```python
#Identificar valores nulos por columnas
valoresnulos=data2.isnull().sum()
valoresnulos
```

**Métodos Sustitucion de valores nulos.**

```python
#Realizamos una copia de dataframe
data3=data2.copy()
```

```python
#Primer metodo de sustitucion de valores nulos
#Sutituir valores nulos con promedio o media
data3['Superficie']=data3['Superficie'].fillna(round(data2['Superficie'].mean(),1))
```

```python
#Comprobamos que ya no tengamos valores nulos
valoresnulos=data3.isnull().sum()
valoresnulos
```

```python
#Segundo metodo de sustitucion de valores nulos
#Sutituir valores nulos con mediana
data3['Precio Venta']=data3['Superficie'].fillna(round(data2['Superficie'].median(),1))
```

```python
#Comprobamos que ya no tengamos valores nulos
valoresnulos=data3.isnull().sum()
valoresnulos
```

```python
#Tercer metodo de sustitucion de valores nulos
#Sutituir valores nulos con una constante que nosotros escogemos
data3['Fecha Alta']=data3['Fecha Alta'].fillna(10)
```

```python
#Comprobamos que ya no tengamos valores nulos
valoresnulos=data3.isnull().sum()
valoresnulos
```

```python
#Cuarto metodo de sustitucion de valores nulos
#Sutituir valores nulos con un string en concreto
data3['Fecha Venta']=data3['Fecha Venta'].fillna("2030-06-19 00:00:00")
```

```python
#Comprobamos que ya no tengamos valores nulos
valoresnulos=data3.isnull().sum()
valoresnulos
```

```python
#Quinto metodo de sustitucion de valores nulos
#Sutituir valores nulos por el valor no nulo siguiente
data3['Tipo']=data3['Tipo'].fillna(method="ffill")
```

```python
#Comprobamos que ya no tengamos valores nulos
valoresnulos=data3.isnull().sum()
valoresnulos
```

```python
#Sexto metodo de sustitucion de valores nulos
#Sutituir valores nulos por el valor no nulo anterior
data3['Vendedor']=data3['Vendedor'].fillna(method="bfill")
```

```python
#Comprobamos que ya no tengamos valores nulos
valoresnulos=data3.isnull().sum()
valoresnulos
```

```python
#Acabamos de sustituir todos los valores nulos
data3['Operación']=data3['Operación'].fillna(method="bfill")
data3['Provincia']=data3['Provincia'].fillna(method="bfill")
data3['Vendedor']=data3['Vendedor'].fillna(10)
```

```python
#Comprobamos que ya no tengamos valores nulos
valoresnulos=data3.isnull().sum()
valoresnulos
```
